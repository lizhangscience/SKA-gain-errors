```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#import seaborn as sns
#sns.set()
```

```python
#Data and plots directories

data_directory = 'datafiles'
plots_directory = 'plots'
```

```python
#Datafiles

#Basic files
datafiles_basic = ['ps_eor_3d.txt', 'ps_fg_point_3d.txt', 'ps_fg_diffuse_3d.txt',
             'ps_noise_3d.txt'] 
datafiles_basic_labels = ['eor', 'fg_point', 'fg_diffuse', 'noise']
```

```python
#Diffuse foreground, uncorrelated gain errors
datafiles_diffuse_uncorrelated = ['ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_uncorrelated_3d_ge_rms_0.01.txt', 
                                'ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_uncorrelated_3d_ge_rms_0.1.txt', 
                                'ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_uncorrelated_3d_ge_rms_1.txt']
datafiles_diffuse_uncorrelated_labels = ['diffuse_uncor_0.01', 'diffuse_uncor_0.1', 'diffuse_uncor_1']

#Diffuse foreground, correlated gain errors
datafiles_diffuse_correlated = ['ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_correlated_3d_ge_rms_0.01.txt', 
                                'ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_correlated_3d_ge_rms_0.1.txt', 
                                'ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_correlated_3d_ge_rms_1.txt']
datafiles_diffuse_correlated_labels = ['diffuse_cor_0.01', 'diffuse_cor_0.1', 'diffuse_cor_1']
```

```python
#Create 'k' column

filename = datafiles_basic[0]
datafile = data_directory + '/' + filename
df = pd.read_csv(datafile, skiprows=3, delim_whitespace=True, usecols=[0], names=['k'])
```

```python
#Read spherically averaged 3D power spectrum for basic signals file

filenames = datafiles_basic
labels = datafiles_basic_labels

for i in range(len(filenames)):
    filename = filenames[i]
    label = labels[i]
    datafile = data_directory + '/' + filename
    df1 = pd.read_csv(datafile, skiprows=3, delim_whitespace=True, usecols=[1], names=[label])
    df = pd.concat([df,df1],axis=1,join='inner')
```

```python
#Finding viable region of operation in (t_coh, sigma) space at the lowest k-value, k=0.0707

delta2_eor = 0.5 #EoR level in mK^2, at k=0.0707

timescale = [15.0, 30.0, 60.0, 120.0, 240.0] 
ge_rms = [0.01, 0.1, 1]

df = pd.DataFrame(columns=['tcoh','sigma','d2'])


for i in range(len(timescale)):
    for j in range(len(ge_rms)):
        filename = 'ps_fg_p_eor_p_ge_m_fg_m_eor_diffuse_correlated_3d_ge_rms_%s_timescale_%s_min.txt'%(str(ge_rms[j]), str(timescale[i]))
        datafile = data_directory + '/' + filename
        df1 = pd.read_csv(datafile, skiprows=3, nrows=1, delim_whitespace=True, usecols=[1], names=['d2'])
        df1['tcoh'] = [timescale[i]]
        df1['sigma'] = [ge_rms[j]]
        df = pd.concat([df,df1])
        
cols = ['tcoh','sigma','d2']
df = df[cols]
```

```python
df_2d = df.pivot('sigma', 'tcoh', 'd2')
df_2d.sortlevel(level=0, ascending=False, inplace=True)
```

```python
import seaborn as sns
sns.set()
from matplotlib.colors import LogNorm

f = sns.heatmap(df_2d,cmap='YlGnBu',annot=True,norm=LogNorm(vmin=df_2d.min(), vmax=df_2d.max()))
f.set(xlabel='$t_{coh}$ [min]',ylabel='$\sigma$')
plt.show()
```
